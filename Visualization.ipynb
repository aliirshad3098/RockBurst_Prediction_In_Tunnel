```python
# feature = 'Duration_days'

# plt.figure(figsize=(12,5))

# # Box Plot
# plt.subplot(1, 2, 1)
# sns.boxplot(x=df[feature], color='skyblue')
# plt.title(f"Boxplot of {feature}")

# # Histogram
# plt.subplot(1, 2, 2)
# sns.histplot(df[feature], kde=True, color='teal', bins=20)
# plt.title(f"Histogram of {feature}")

# plt.tight_layout()
# plt.show()
# # Skewness
# sk = df[feature].skew()
# print(f"Skewness of {feature}: {sk}")
```

```python
# Function to draw plots and return skewness
def draw_graphs(feature, message):
    plt.figure(figsize=(12,5))

    # Box Plot
    plt.subplot(1, 2, 1)
    sns.boxplot(x=df[feature], color='skyblue')
    plt.title(f"Boxplot of {feature} {message}")

    # Histogram
    plt.subplot(1, 2, 2)
    sns.histplot(df[feature], kde=True, color='teal', bins=20)
    plt.title(f"Histogram of {feature} {message}")

    plt.tight_layout()
    plt.show()

    # Skewness
    sk = df[feature].skew()
    print(f"Skewness of {feature} {message}: {sk:.3f}")
    return sk
```

```python
# Features to process
features = df.columns.drop(["Intensity_Level_encoded","Intensity_Level"])
new_features = []

# Loop for each feature
for feature in features:
    sk = draw_graphs(feature, "Before Processed")  # draw before transformation
    
    if sk > 0.5:
        df[f"{feature}_log"] = np.log1p(df[feature])
        new_features.append(f"{feature}_log")
        print(f"Applied log1p transformation on {feature}")
        
    elif sk < -0.5:
        df[f"{feature}_sqr"] = np.square(df[feature])
        new_features.append(f"{feature}_sqr")
        print(f"Applied square transformation on {feature}")
        
    else:
        print(f"No transformation applied on {feature}")
        continue
    
    # Show after transformation
    new_feature = new_features[-1]  # last added feature
    draw_graphs(new_feature, "After Processed")

print("Newly created features:", new_features)
```

```python

n_features = len(features)

# Grid layout: 3 columns (adjust as needed)
n_cols = 2
n_rows = (n_features + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 5))
axes = axes.flatten()

for i, feature in enumerate(features):
    sns.boxplot(x='Intensity_Level', y=feature, data=df, palette='Set1', ax=axes[i],hue="Intensity_Level",legend=False)
    axes[i].set_title(f"{feature} by Intensity Level", fontsize=10)
    axes[i].set_xlabel("Intensity Level")
    axes[i].set_ylabel(feature)

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.suptitle("Feature Distribution by Rockburst Intensity Level", fontsize=16, y=1.02)
plt.show()
```
