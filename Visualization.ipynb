## The distribution is not uniform we need to make them balance in the preprocessing phase

```python
df.columns[9:]
```

```python
features = ['Energy_Unit_log', 'Energy_density_sqr', 'Volume_sqr',
       'Event_freq_unit_per_day_log', 'Energy_Joule_per_day_sqr',
       'Volume_m3_per_day_sqr','Duration_days']  # All features except the target
```

```python

n_features = len(features)

# Grid layout: 3 columns (adjust as needed)
n_cols = 2
n_rows = (n_features + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 5))
axes = axes.flatten()

for i, feature in enumerate(features):
    sns.boxplot(x='Intensity_Level', y=feature, data=df, palette='Set1', ax=axes[i],hue="Intensity_Level",legend=False)
    axes[i].set_title(f"{feature} by Intensity Level", fontsize=10)
    axes[i].set_xlabel("Intensity Level")
    axes[i].set_ylabel(feature)

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.suptitle("Feature Distribution by Rockburst Intensity Level", fontsize=16, y=1.02)
plt.show()
```

```python
columns=features
columns.append('Intensity_Level_encoded')
plt.figure(figsize=(10,6))
sns.heatmap(df[columns].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix Between Features")
plt.xticks(rotation=45)
plt.show()
```

```python
# Create the pairplot
plot = sns.pairplot(df, vars=features, hue='Intensity_Level', palette='husl')

# Add a super title
plot.fig.suptitle("Pairwise Relationships of Rockburst Features by Intensity Level", 
                  fontsize=14, y=1.02)  # y=1.02 moves the title above the plot

plt.show()
```

```python
# Grid layout
n_cols = 2
n_rows = (n_features + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 5))
axes = axes.flatten()

for i, feature in enumerate(features):
    sns.histplot(df[feature], kde=True, bins=20, color='teal', ax=axes[i])
    axes[i].set_title(f"Distribution of {feature}", fontsize=20)
    axes[i].set_xlabel(feature,fontsize=15)
    axes[i].set_ylabel("Frequency",fontsize=15)

# Remove any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.suptitle("Feature Distribution by Rockburst Intensity Level", fontsize=16, y=1.02)
plt.show()
```
